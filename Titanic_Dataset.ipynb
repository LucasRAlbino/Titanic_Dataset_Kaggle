- ### Treinar um modelo utilizando arvores de decisão

```python
dataset.head()
```

```python
# Separar entre sexo, idade e classe X
# Sobreviveu Y

x = dataset[['Sex', 'Age', 'Pclass']].copy()
y = dataset['Survived']

# Na coluna 'Sex', colocaremos 0 para masculino e 1 para feminino

x['Sex'] = x['Sex'].map({"male": 0, "female": 1})

# A arvore vai tratar como um valor diferente de todos os outros
x = x.fillna(-1)
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.3, random_state = 42)

modelo = DecisionTreeClassifier(random_state = 0, min_samples_leaf=4)
modelo.fit(xtrain, ytrain)

p = modelo.predict(xtest)
accuracy_score(ytest, p)
```
